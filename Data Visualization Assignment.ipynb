## Data Visualization

In this assignment students have to transform iris data into 3 dimensions  and plot a 3d chart with transformed dimensions and colour each data  point with specific class. 
 


```python
import numpy as np 
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D 
from sklearn import decomposition  
from sklearn import datasets 
from sklearn.decomposition import PCA
```

```python
df= datasets.load_iris()   #loading dataset
```

```python
#As we have three components
pca = PCA(n_components = 3)
pca.fit(df.data)
x_pca=pca.transform(df.data)
y=df.target 
```

```python
x_pca.shape
```

```python
fig = plt.figure(1, figsize=(14,10))
dim=Axes3D(fig, elev = -140, azim =130)

dim.scatter(x_pca[:,0],x_pca[:,1],x_pca[:,2],c = y, cmap = 'plasma')

dim.set_xlabel("First Dimention", fontsize=12)
dim.set_ylabel("Second Dimension",fontsize=12)
dim.set_zlabel("Third Dimension", fontsize=12)
dim.set_title("Iris Dataset Dimensions", fontsize=18)

plt.show()
```
